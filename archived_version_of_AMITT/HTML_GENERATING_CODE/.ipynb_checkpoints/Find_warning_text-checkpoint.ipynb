# Find warning text in AMITT generated files

```python
import os
import glob

warntext = 'DO NOT EDIT ABOVE THIS LINE - PLEASE ADD NOTES BELOW'
warnlen = len(warntext)

objecttypes = {'phase','tactic','technique','task','incident',
    'counter','metatechnique','actortype',#'resource',
    #'responsetype',#'detection'
}

foundtext = []
for objecttype in objecttypes:

    objecttypeplural = objecttype + 's'
    objecttypedir = '../{}'.format(objecttypeplural)
    if not os.path.exists(objecttypedir):
        print("directory {} doesn't exist".format(objecttypedir))
        continue

    for datafile in glob.glob(objecttypedir + '/*.md'):
        with open(datafile) as f:
            filetext = f.read()
        warnpos = filetext.find(warntext)
        endpos = warnpos+len(warntext)
        if endpos != len(filetext):
            objectname = datafile[datafile.rfind('/')+1:-3]
            print('warning text in {} {}'.format(objecttype, objectname)) 
            foundtext += [[objecttype, objectname, filetext[endpos:]]]
foundtext
```

```python
import pandas as pd
import csv

dffound = pd.DataFrame(foundtext)
dffound.to_csv('TEST_foundtext.csv', quoting=csv.QUOTE_NONNUMERIC,
               index=False)
dffound
```
